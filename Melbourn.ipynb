```python
import pandas as pd

# Define the file path
file_path = 'data/Melbourn_housing.csv'

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to verify
print(df.head())
```

```python
# Remove the 'Method' column
df.drop(columns=['Method'], inplace=True)
```

```python
df
```

```python
# Divide the 'Price' column by 1000 and create a new column 'Price K$'
df['Price K$'] = df['Price'] / 1000

# Drop the old 'Price' column
df.drop(columns=['Price'], inplace=True)
```

```python
df
```

```python
# Divide the 'Landsize' column by 4046.86 to convert to acres and create a new column 'Landsize (acres)'
df['Landsize (acres)'] = df['Landsize'] / 4046.86
```

```python
df
```

```python
# Function to split the address into number and name parts
def split_address(address):
    import re
    match = re.match(r"(\d+[\w\W]*?)\s(.*)", address)
    if match:
        return match.groups()
    else:
        return address, ""

# Apply the function to the 'Address' column and create new columns
df[['Address Number', 'Address Name']] = df['Address'].apply(split_address).apply(pd.Series)
```

```python
# Export the DataFrame to a CSV file called 'final.csv'
df.to_csv('final_Melbourn.csv', index=False)
```

```python
# Function to calculate Haversine distance in meters
def haversine(lat1, lon1, lat2, lon2):
    from math import radians, sin, cos, sqrt, atan2
    
    # Radius of the Earth in meters
    R = 6371000
    
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    
    # Compute differences
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    # Haversine formula
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    # Distance in meters
    distance = R * c
    return distance

# Coordinates of the reference point
ref_lat = -37.80140
ref_lon = 144.99580

# Calculate distance for each row and create a new column 'Distance (m)'
df['Distance (m)'] = df.apply(lambda row: haversine(ref_lat, ref_lon, row['Lattitude'], row['Longtitude']), axis=1)
```

```python
import plotly.express as px

# 📌 Ensure required columns exist and clean data
if {"Lattitude", "Longtitude", "Type"}.issubset(df.columns):
    df = df.dropna(subset=["Lattitude", "Longtitude", "Type"])  # Remove rows with NaN values

    # 🔹 Convert to numeric to avoid errors
    df["Lattitude"] = pd.to_numeric(df["Lattitude"], errors="coerce")
    df["Longtitude"] = pd.to_numeric(df["Longtitude"], errors="coerce")

    # 🔹 Define a color mapping for the 'Type' column
    color_map = {"u": "blue", "t": "green", "h": "red"}  # Customize colors

    # 🔹 Create the Plotly scatter map
    fig = px.scatter_mapbox(
        df,
        lat="Lattitude",
        lon="Longtitude",
        color="Type",
        color_discrete_map=color_map,
        hover_data=["Type"],  # Show type on hover
        title="Melbourne Housing Locations by Type",
        zoom=10
    )

    # 🔹 Use an OpenStreetMap basemap
    fig.update_layout(
        mapbox_style="open-street-map",
        mapbox_center={"lat": df["Lattitude"].mean(), "lon": df["Longtitude"].mean()},
        margin={"r":0,"t":40,"l":0,"b":0}
    )

    # 🔹 Show the map
    fig.show()

else:
    print("⚠️ The required columns 'Lattitude', 'Longtitude', or 'Type' are missing in the DataFrame.")
```
