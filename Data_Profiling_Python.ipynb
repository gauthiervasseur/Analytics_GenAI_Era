# Set your file path

```python
# Define the file path
file_path = 'data/Melbourn_housing.csv'
```

# YData Profiling

```python
#Install the below libaries before importing
import pandas as pd
from ydata_profiling import ProfileReport

profile = ProfileReport(pd.read_csv(file_path), title="Your Dataset Report", explorative=True, correlations={
    "pearson": {"calculate": True},
    "spearman": {"calculate": True},
    "kendall": {"calculate": True},
    "phi_k": {"calculate": True},
    "cramers": {"calculate": True}
})

#EDA using pandas-profiling
# profile = ProfileReport(pd.read_csv(file_path), explorative=True)

#Saving results to a HTML file
profile.to_file("Data_profile.html")

# Display the report inline in a Jupyter Notebook
profile.to_notebook_iframe()

```
